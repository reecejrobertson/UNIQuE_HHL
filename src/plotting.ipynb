```python
import ast
import matplotlib
from matplotlib import pyplot as plt
```

```python
matplotlib.rc('text', usetex=True)
matplotlib.rc('font', family='serif')
plt.rcParams['figure.dpi'] = 150
plt.rcParams['axes.axisbelow'] = True

BIG_TEXT = 12
MEDIUM_TEXT = 10
SMALL_TEXT = 8
COLUMN_WIDTH = 3.5
```

```python
with open('data/counts.txt') as file:
    data = ast.literal_eval(file.read())
data
```

```python
for experiment, counts in data.items():
    x = list(counts.keys())
    y = list(counts.values())
    fig = plt.figure(figsize=(COLUMN_WIDTH, COLUMN_WIDTH/2))
    plt.bar(x, y)
    for i in range(len(y)):
        plt.annotate(
            str(y[i]),
            xy=(x[i], y[i]),
            xytext=(0, 5),
            textcoords="offset points",
            ha='center',
            fontsize=SMALL_TEXT
        )
    plt.grid(which='major', color='0.85', linestyle=':')
    plt.tick_params(which='major', labelsize=SMALL_TEXT)
    plt.xlabel('Result', fontsize=MEDIUM_TEXT)
    plt.ylabel('Frequency', fontsize=MEDIUM_TEXT)
    plt.ylim(0, max(y) + max(y)/5)
    plt.savefig(f'figs/{experiment}.png')
    plt.show()
    plt.close()
```
